#### **Missing Values:** Correlation between missing values and compliance or risk

---
> The last thing about missing values we need to find before moving on to handling them: Is there any correlation between the missing values and the compliance or risk? Because if there is, then something about the missing data can affect the compliance or risk and we want to know what before handling missing values. Perhaps the number of missing values negatively correlates with the compliance and positively correlates with risk.
---

```python
exclude_cols = ['dept_id', 'dept_name', 'dept_category', 'division']
analytic_cols = departments.columns.difference(exclude_cols)

# Fraction of missing values per row
row_missing_rate = departments[analytic_cols].isna().mean(axis=1)

compliance_col = 'compliance_score_final'
risk_col = 'overall_risk_score'

correlation_compliance = row_missing_rate.corr(departments[compliance_col])
correlation_risk = row_missing_rate.corr(departments[risk_col])

print(f"Correlation between missingness and compliance_score:final: {correlation_compliance:.3f}")
print(f"Correlation between missingness and overall_risk_score: {correlation_risk:.3f}")
```

> Correlation with compliance (0.033) ‚Üí very low, meaning the amount of missing data does not systematically affect compliance scores.
>
> Correlation with risk (0.006) ‚Üí essentially zero, so missing data is not predictive of risk either.
>
> In other words, based on this dataset:
>- Departments with more missing values are not inherently riskier.
>- Missingness seems to be mostly structural (like Operational_Compliance departments not submitting data), not related to the outcome variables.

```python
# Merge departments with high_risk_departments on dept_id, keeping both versions
comparison_table = pd.merge(
    departments, 
    high_risk_departments, 
    on='dept_id', 
    how='inner', 
    suffixes=('_orig', '_highrisk')
)

# Display the new comparison table
comparison_table.head()
```

```python
comparison_results = {}

# Only compare analytical/important columns (exclude dept_id, dept_name, dept_category, division)
cols_to_check = [col for col in departments.columns if col not in ['dept_id', 'dept_name', 'dept_category', 'division']]

for col in cols_to_check:
    orig = comparison_table[f"{col}_orig"]
    clean = comparison_table[f"{col}_highrisk"]
    
    imputed = ((orig.isna()) & (~clean.isna())).sum()       # NaN -> value
    untouched = ((orig.isna()) & (clean.isna())).sum()      # NaN -> NaN
    unchanged = ((~orig.isna()) & (~clean.isna()) & (orig == clean)).sum()  # value -> same value
    changed = ((~orig.isna()) & (~clean.isna()) & (orig != clean)).sum()    # value -> changed value
    
    comparison_results[col] = {
        'imputed_count': imputed,
        'untouched_count': untouched,
        'unchanged_count': unchanged,
        'changed_count': changed
    }

# Convert to DataFrame for display
comparison_df = pd.DataFrame(comparison_results).T
comparison_df = comparison_df.sort_values('imputed_count', ascending=False)
display(comparison_df)
```

> 1) We see that operational or critical metrics (like reporting, compliance, audits) were imputed or corrected (see later outliers check).
> 2) Less critical or harder-to-guess fields (age, violations, oversight body) were left missing.
> 3) It is okay to leave some data missing.
>
> We derive this pattern so far:
>- Critical analytical columns ‚Üí impute if missing or correct if strange.
>- Non-critical descriptive columns ‚Üí it‚Äôs okay to leave some missing if unsure.

### Work on the main departments table, drop rows with missing division, and drop the dept_name column

```python
# Count how many rows have division = NaN
missing_division_count = departments['division'].isna().sum()

print("Number of rows with missing 'division':", missing_division_count)

# Optional: quickly look at a few of these rows
# to remind ourselves what they look like
print("\nExample rows with missing 'division':")
display(departments[departments['division'].isna()].head())


# Drop rows where division is missing
# Create a cleaned copy so we don't overwrite the original immediately
departments_clean = departments.copy()

# Keep only rows where division is NOT NaN
departments_clean = departments_clean[~departments_clean['division'].isna()]

print("\nShape BEFORE dropping missing division rows:", departments.shape)
print("Shape AFTER  dropping missing division rows:", departments_clean.shape)

#Drop the dept_name column
# First, check that the column exists (to avoid errors if already dropped)
if 'dept_name' in departments_clean.columns:
    departments_clean = departments_clean.drop(columns=['dept_name'])
    print("\nColumn 'dept_name' has been dropped from departments_clean.")
else:
    print("\nColumn 'dept_name' is not present (already dropped earlier).")

#Quick sanity check of the cleaned table
print("\nColumns in departments_clean after cleaning:")
print(departments_clean.columns.tolist())

print("\nPreview of the cleaned departments table:")
display(departments_clean.head())
```

### Step 1: Cleaning the departments table by dropping useless rows and columns
We apply the first cleaning action in our main table (departments): We dropped all rows where the division was missing, as we had earlier discovered a strong pattern. From that pattern, we saw that if the division is missing, then all other categorical values are also missing (only dept_id, dept_name, and dept_category contain values). These rows have 100% missing data for the variables we would use for modelling, and these rows are therefore non-usable. In addition, we found that every single row with a missing division belongs to the Operational Compliance category, meaning that a large subset of this category is useless and cannot be used in analysis. 
### Step 2: Drop the dept_name column
We decided earlier that dept_name is a purely descriptive label and that it does not carry any analytical information that is useful for predicting compliance or risk. Only dept_id is a unique and reliable identifier for departments. For this reason, we drop dept_name from the departments to:
1. Reduce noise,
2. Avoid leaking textual labels into the model,
3. Keep only stable identifiers.

Now, departments contain only rows with known division, all empty Operational Compliance rows are removed, and dept_name is dropped while dept_id remains as the reliable key.


### Impute critical analytical columns

```python
import numpy as np

#Define the critical columns to impute
# Numeric compliance/risk metrics
numeric_impute_cols = [
    "reporting_lag_days",
    "training_hours_quarterly",
    "reporting_gaps_annual",
    "audit_score_q1",
    "audit_score_q2"
]

# Binary/bool-like operational indicators
binary_impute_cols = [
    "digital_systems",
    "onboarding_program",
    "improvement_commitment"
]

#Impute NUMERIC columns with the MEDIAN
for col in numeric_impute_cols:
    
    # Check if column exists
    if col not in departments_clean.columns:
        print(f" Column '{col}' not found ‚Äî skipping.")
        continue
    
    median_val = departments_clean[col].median()
    
    # Fill missing values with the median
    departments_clean[col] = departments_clean[col].fillna(median_val)
    
    print(f" Imputed missing values in '{col}' using MEDIAN = {median_val}")

#Impute BINARY columns with the MODE
for col in binary_impute_cols:

    if col not in departments_clean.columns:
        print(f" Column '{col}' not found ‚Äî skipping.")
        continue
    
    mode_val = departments_clean[col].mode().iloc[0]  # Mode returns a Series
    
    # Fill missing values with the mode
    departments_clean[col] = departments_clean[col].fillna(mode_val)
    
    print(f" Imputed missing values in '{col}' using MODE = {mode_val}")


# Quick verification
print("\n Checking remaining missing values in imputed columns:")
print(departments_clean[numeric_impute_cols + binary_impute_cols].isna().sum())

print("\n All critical analytical columns are now imputed.\n")
```

### Why do we impute? - Because ML cannot work with missing values.
Instead of deleting rows because something is missing, we fill the empty value with something practical. From our earlier analysis, in the 'high_risk_departments_table', we identified a set of critical operational and compliance indicators that must be imputed when missing. These columns are essential because:
1. They describe how well a department reports, audits, trains, and follows procedures,
2. They appear in the current high_risk table, where missing values were already filled,
3. They directly influence risk levels, operational health, and compliance performance.

We used median imputation because for numeric values, the median is robust against outliers and it prevents central tendency, without being skewed by extreme values.
We used mode imputation for binary values because it is the most common value to fill the binary missing values.

### Flag missing values in descriptive fields

```python
#Choose the columns where we want to ADD "missing" flags
# These are descriptive / hard-to-guess fields.
flag_cols = [
    "dept_age_years",
    "violations_past_3years",
    "primary_function",
    "secondary_function",
    "oversight_body",
    "location_type",
    "team_size",
    "reporting_structure",
]

#Create a new *_missing column for each of them
for col in flag_cols:
    # Only do this if the column actually exists in the dataframe
    if col not in departments_clean.columns:
        print(f" Column '{col}' not found in departments_clean ‚Äî skipping.")
        continue

    # Name of the new flag column, e.g. "dept_age_years_missing"
    missing_col = col + "_missing"

    # .isna() gives True where value is NaN, False otherwise
    # .astype(int) converts True ‚Üí 1, False ‚Üí 0
    departments_clean[missing_col] = departments_clean[col].isna().astype(int)

    print(f" Created flag column '{missing_col}' (1 = missing, 0 = not missing).")

#Quick check: how many missing entries did we flag?
print("\n Number of missing values flagged in each column:")
flag_cols_existing = [c for c in flag_cols if c in departments_clean.columns]
missing_flag_names = [c + "_missing" for c in flag_cols_existing]

# Sum of 1s in each *_missing column = number of missing values in original column
print(departments_clean[missing_flag_names].sum())

print("\n Flagging complete. Descriptive columns remain with NaNs,")
print(" but we now have *_missing columns telling us where data was missing.")
```

### Why do we not impute these descriptive values?
Because:
1. There is no clear pattern to predict them,
2. They are not as critical for compliance prediction,
3. Imputing them could introduce misleading information.
### Missing values cause problems for ML mode:
That is why we flag the columns with missing values, so the ML model won't error, the model learns that misleadingness by itself might be meaningful. 
This way we tell the model: this row had missing info here.

### Outliers check

```python
import numpy as np

#Numeric outliers using IQR
numeric_outlier_cols = [
    "reporting_lag_days",
    "training_hours_quarterly",
    "reporting_gaps_annual",
    "violations_past_3years",
    "audit_score_q1",
    "audit_score_q2",
    "audit_score_q3",
    "audit_score_q4",
    "compliance_score_final",
    "overall_risk_score",
    "risk_exposure_operational",
    "risk_exposure_financial",
    "operational_health_index",
    "interdept_collaboration_score",
    "resource_availability_score",
    "external_interactions_frequency",
    "dept_age_years",
]

print("Numeric outliers (IQR method):\n")

outlier_summary = []

for col in numeric_outlier_cols:
    if col not in departments_clean.columns:
        print(f"Column '{col}' not found ‚Äî skipping.")
        continue
    
    # Drop NaNs for quantile calculation
    col_data = departments_clean[col].dropna()
    if col_data.empty:
        print(f"Column '{col}' has no data ‚Äî skipping.")
        continue
    
    Q1 = col_data.quantile(0.25)
    Q3 = col_data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Boolean mask for outliers
    is_outlier = (departments_clean[col] < lower_bound) | (departments_clean[col] > upper_bound)
    n_outliers = is_outlier.sum()
    
    outlier_summary.append((col, n_outliers, float(lower_bound), float(upper_bound)))
    print(f"- {col}: {n_outliers} potential outliers (lower={lower_bound:.2f}, upper={upper_bound:.2f})")

# If you want a tidy summary table:
import pandas as pd
outlier_df = pd.DataFrame(outlier_summary, columns=["column", "n_outliers", "lower_bound", "upper_bound"])

print("\n Outlier summary table:")
display(outlier_df)


#Check binary variables (should be only 0 or 1)
binary_cols = [
    "digital_systems",
    "onboarding_program",
    "improvement_commitment",
    "executive_support",
    "external_consulting",
    "engagement_programs",
    "remediation_plan_active",
    "external_partnerships",
]

print("\n Binary variables: values that are NOT 0 or 1\n")

for col in binary_cols:
    if col not in departments_clean.columns:
        print(f"Column '{col}' not found ‚Äî skipping.")
        continue
    
    # Find all unique values that are not 0, 1, or NaN
    invalid_vals = departments_clean[col].dropna().unique()
    invalid_vals = [v for v in invalid_vals if v not in [0, 1]]
    
    if len(invalid_vals) == 0:
        print(f"- {col}: OK (only 0/1/NaN)")
    else:
        print(f"- {col}: unexpected values {invalid_vals}")



# Check team_size categories
print("\nTeam size categories:\n")

if "team_size" in departments_clean.columns:
    print(departments_clean["team_size"].value_counts(dropna=False))
else:
    print("Column 'team_size' not found.")



# Logical check on dept_age_years
print("\n Logical check: department age\n")

if "dept_age_years" in departments_clean.columns:
    negative_ages = (departments_clean["dept_age_years"] < 0).sum()
    very_large_ages = (departments_clean["dept_age_years"] > 100).sum()
    
    print(f"- Rows with dept_age_years < 0: {negative_ages}")
    print(f"- Rows with dept_age_years > 100: {very_large_ages}")
    
    # Optional: see some examples
    if negative_ages > 0:
        print("\nExamples of negative ages:")
        display(departments_clean[departments_clean["dept_age_years"] < 0].head())
    if very_large_ages > 0:
        print("\nExamples of very large ages (> 100):")
        display(departments_clean[departments_clean["dept_age_years"] > 100].head())
else:
    print("Column 'dept_age_years' not found.")


print("\n Outlier checking step finished. We have only flagged and reported outliers.")
print(" Next step is to decide case-by-case whether to keep, cap, correct, or drop them.")
```

### Outliers: Looking for strange/impossible values
We focus on the columns that are very important for compliance and risk indicators (numeric).
For these columns, we use the IQR rule: anything below or above IQR is a potential outlier. Here, we do not immediately delete or change them; we first report how many there are. 
For the binary variables, any value that is not 0 or 1 is treated as a logical error. Here, we can see that the binary data is clean and valid.
The team size categories are clean: only small, medium and large. 
The department age (how old is the department) follows the logic, is not < 0 (0 years old), nor bigger than 100 (100 years old). Meaning the department's age is reasonable. 
We have successfully flagged the outliers here!
